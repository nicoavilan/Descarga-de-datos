# Descargar datos de Wikipedia

Es posible cargar datos que están en las tablas de Wikipedia.

Una referencia útil es [pypi.org/project/wikipedia](https://pypi.org/project/wikipedia/).


En este ejemplo se van a cargar los datos de la página [Anexo:Localidades de Bogotá](https://es.wikipedia.org/wiki/Anexo:Localidades_de_Bogot%C3%A1).

Inicialmente se instala la librería de Wikipedia

```python
!pip install wikipedia
```

Cargo la librería [Pandas](https://pandas.pydata.org/).

```python
import pandas as pd
```

La página [Anexo:Localidades de Bogotá](https://es.wikipedia.org/wiki/Anexo:Localidades_de_Bogot%C3%A1) tiene dos tablas de datos. Se cargan con la siguiente instrucción

```python
datos = pd.read_html("https://es.wikipedia.org/wiki/Anexo:Localidades_de_Bogot%C3%A1")
datos
```

La primera tabla de datos quedó en el primer DataFrame, que es la componente 0 del archivo que se cargó.

```python
datos[0]
```

La segunda tabla de datos quedó en el tercer DataFrame, que es la componente 2 del archivo que se cargó.

```python
datos[2]
```
