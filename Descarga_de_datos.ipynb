## Descarga de un archivo csv con caracteres latinos (ñ, á, é, í, ó, ú, etc)


Se van a leer los datos del archivo [OSB_EnfTransm-COVID-19 April 2020.csv](https://github.com/nicoavilan/Bases-de-datos/blob/main/OSB_EnfTransm-COVID-19%20April%202020.csv) para hacerlo se toma la dirección de los datos crudos (raw) [OSB_EnfTransm-COVID-19 April 2020.csv (datos crudos)](https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/OSB_EnfTransm-COVID-19%20April%202020.csv)

Si los datos están separados con ; (punto y coma) se debe especificar en la lectura de los datos usando la instrucción `sep= ";"`.  

```python
url = "https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/OSB_EnfTransm-COVID-19%20April%202020.csv"
df1 = pd.read_csv(url,encoding = "latin-1", sep= ";" )
df1
```
